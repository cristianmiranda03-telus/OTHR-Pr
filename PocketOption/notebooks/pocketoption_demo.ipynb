```python
import sys
from pathlib import Path

ROOT = Path("..").resolve()
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from dotenv import load_dotenv

load_dotenv(ROOT / ".env")
```

```python
from pocketoption_connector import PocketOption

ASSET = "EURUSD_otc"

with PocketOption.session(default_asset=ASSET, load_dotenv=False) as po:
    print("balance:", po.balance)
    print("snapshot keys:", list(po.account_snapshot().keys()))
    print("last candle:", po.last_candle())
```

```python
with PocketOption.session(default_asset=ASSET, load_dotenv=False) as po:
    candles = po.candles(count=15, timeframe=60)
    print("candles:", len(candles))
```

```python
with PocketOption.session(default_asset=ASSET, load_dotenv=False) as po:
    try:
        print(po.assets()[:12])
    except Exception as exc:
        print("Asset list unavailable:", exc)
```
